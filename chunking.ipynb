```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
try:
    with open("./results/extracted_texts.txt", "r") as file:
        content = file.read()
except FileNotFoundError:
    print("File is not available")
```

```python
text_splitter = RecursiveCharacterTextSplitter(separators=["\n","\n\n","."," "], chunk_size=150, chunk_overlap=25)
texts = text_splitter.split_text(content)
```
